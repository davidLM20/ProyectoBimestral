```python
#se importa todas las librerias necesarias
import pandas as pd # se importa la libreria pandas necesaria para el manejo de archivos
import numpy as np # esta libreria se necesaria para representar la informacion que esta vacia
from bokeh.plotting import figure # se importa este paquete de la libreria bokeh para hacer uso de las respectivas graficas
from bokeh.palettes import * # se importa este paquete para hacer uso de los colores
from bokeh.io import output_file, show , output_notebook 
output_notebook() # se genera para la impresion de graficas

datos = pd.read_csv("data/metadatos.csv", sep=";", encoding = "ISO 8859-1") # se lee el archivo csv, a la par se genera un split

```

```python
datos # se muestra la informacion del csv
```

```python
datos.groupby(['Provincia'])['Provincia'].count() # se agrupa toda la informacion dependiendo la llave que se ingrese, a su vez se genera un conteo
```

```python
#se genera una lista esta lista posterior mente servira como el eje x de la grafica
provincias = ['Azuay', 'Bolivar', 'Carchi', 'Cañar', 'Chimborazo', 'Cotopaxi', 'El Oro', 'Esmeraldas','Guayas','Imbabura','Loja','Los_Rios','Manabi','Morona_Santiago','Napo','Orellana','Pastaza','Pichincha','Santa_Elena','Santo_Domingo_De_Los_Tsachilas','Sucumbios','Tungurahua','Zamora_Chinchipe']
```

```python
# se genera un contador para el eje y, a su vez hace un conteo por cada uno de las provincias 
contador = datos.groupby(['Provincia'])['Provincia'].count()
```

```python
output_file("Provincias.html") # abre una pestania para presentar el grafico 
p = figure(x_range=provincias, plot_height=350,plot_width = 2600, title="Numero de muertes por Provincia",toolbar_location=None, tools="") # en esta linea se llama la figura y se le da los parametros
color = ['#225ea8', '#41b6c4', '#a1dab4', '#ffffcc','#CD853F','#FF4500','#393b79', '#5254a3', '#6b6ecf', '#9c9ede''#440154', '#30678D', '#35B778', '#FDE724','#FFD700','#ADFF2F','#00BFFF','#FF1493','#DC143C','#8A2BE2','#FFA07A','#FF4500','#FF8C00','#DA70D6'] 
p.vbar(x=provincias, top=contador, width=0.9, color=color, line_color = "#000000") # se da los parametros de las barras

p.xgrid.grid_line_color = None # color de las lineas posteriores
p.y_range.start = 0
p.yaxis.axis_label = "Numero De Muertes" #legendas de y
p.xaxis.axis_label = "Provincias" #legenda de x
show(p)
```
